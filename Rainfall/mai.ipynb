```python
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("austin_weather_final.csv")
data
```

```python
x = data.drop(['PrecipitationSumInches'], axis=1)
```

```python
y = data['PrecipitationSumInches']
```

```python
y= y.values.reshape(-1,1)
```

```python
y
```

```python
day_index = 798
days = list(range(len(y)))
days
```

```python
clf = LinearRegression()
clf.fit(x,y)
```

```python
print("The precipitation trend graph:")
plt.scatter(days, y, color = 'g')
plt.scatter(days[day_index], y[day_index], color = 'r')
plt.title("Precipitation level")
plt.xlabel("Days")
plt.ylabel("Precipitation in inches")
plt.show()
x_vis = x.filter(['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH'])
```

```python
# run this before plotting
print("len(days):", len(days))
print("x_vis.shape:", x_vis.shape)
print("day_index:", day_index)
# check sample sizes you will plot
print("first 100 days length:", min(100, len(days)))
print("first100 column length:", len(x_vis.iloc[:min(100, len(x_vis)), 0]))
```

```python
print(x_vis.shape)
print(x_vis.columns.tolist())
print("sample values (first col):", x_vis.iloc[:5,0].tolist())
print("n, m, day_index:", len(x_vis), min(100,len(x_vis)), day_index)
```

```python
import matplotlib.pyplot as plt
print("The precipitation vs attributes trend graph:")
plt.figure(figsize=(18, 12))
for i in range(x_vis.columns.size):
    plt.subplot(3, 2, i + 1)
    plt.scatter(days[:100], x_vis[x_vis.columns.values[i]][:100], color='g')
    plt.scatter(days[day_index],x_vis[x_vis.columns.values[i]][day_index], color='r')
    plt.title(x_vis.columns.values[i])
plt.show()
```
